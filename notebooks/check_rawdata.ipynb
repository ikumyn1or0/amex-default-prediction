# Exploring Train Data

```python
train_sample = pd.read_csv("../input/train_data_sample.csv", nrows=100000)
train_sample.head()
```

```python
print(f"rows (sampled): {train_sample.shape[0]}")
print(f"duplicated rows: {train_sample.duplicated().sum()}")
print(f"columns: {train_sample.shape[1]}")
```

```python
train_info = pd.DataFrame(data=train_sample.dtypes, index=train_sample.columns, columns=["dtypes"])
train_info["count"] = train_sample.count()
train_info["NAs"] = train_sample.isna().sum()
train_info["length"] = train_info["count"] + train_info["NAs"]
train_info["nunique"] = train_sample.nunique()
train_info
```

## [WIP]check column

```python
features = ["B_", "D_", "P_", "R_", "S_"]
feature = features[1]
columns = list(train_sample.columns)
check_columns = [feature+str(int(c[2:])) for c in columns if feature in c]
print(check_columns)
```

```python
# column = "D_65"
column = check_columns[39]
print(column)
print(f"NAs: {train_sample[column].isna().sum()}")
figure = plt.figure(figsize=(25, 10))
ax1 = figure.add_subplot(2, 1, 1)
sns.histplot(data=train_sample, x=column, bins=100)
ax2 = figure.add_subplot(2, 1, 2)
xlim = [0, train_sample[column].max()/10]
# xlim = [0., 10/38]
sns.histplot(data=train_sample[["customer_ID", column]].query(f"{xlim[0]} <= {column} <= {xlim[1]}"), x=column, bins=1000)
plt.xlim(xlim)
plt.ylim([0, 100])
```
